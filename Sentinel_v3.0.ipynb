```python
from sklearn import svm

def predict(File):  
    """delete the 'Time', 'Open', 'High', 'Low' and 'Volume'
    Columns from the dataset as we don't need them"""
    File.drop(File.loc[:, ["Time", "Open", "High", "Low", "Volume"]], inplace=True, axis=1)


    """Converting the 'Date' column from a string to a datetime"""
    File['Date'] = pd.to_datetime(File['Date'])
    
    """Assigning new columns for our next calculations"""
    #File['Year'] = File.Date.dt.year
    #File['Week_day_#'] =File.Date.dt.weekday
    #File['Day_name'] = File.Date.dt.weekday_name
    #File['Week_#'] = File.Date.dt.weekofyear
    File['Returns'] = np.log(File.Close/File.Close.shift(1))
   
    # we are adding the lag columns based off of the direction 252 years again
    for lag in range(0, lags):
        col = 'lag_%s' % (lag+1)
        File[col] = File['Returns'].shift(lag*252)
        cols.append(col)
        
    # to Drop off "NaN" values    
    File.dropna(inplace=True)
    
    # to describe the direction of the market with "1" (as Bull), "-1" (as Bear) and "0" (as Neutral)
    File['Direction'] = np.sign(File['Returns'])
    
    #Create a model to use in our prediction
    model = svm.SVC(C=100)
    model.fit(np.sign(File[cols]),np.sign(File.Returns))
    return File
   
```
