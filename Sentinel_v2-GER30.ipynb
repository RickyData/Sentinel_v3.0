```python
import pandas as pd
import numpy as np
```

```python
path = 'GER301440.csv'
```

```python
eu = pd.read_csv(f'{path}', names=["Date", "Time", "Open", "High", "Low", "Close", "Volume"])
```

```python
def clean(File):  
    """We need only 'Date' and 'Close' Columns from the dataset """
    #File.drop(File.loc[:, ["Time", "Open", "High", "Low", "Volume"]], inplace=True, axis=1)
    File = File.loc[:,['Date', 'Close']]


    """Converting the 'Date' column from a string to a datetime"""
    File['Date'] = pd.to_datetime(File['Date'])
    
    """Assigning new columns for our next calculations"""
    File['Year'] = File.Date.dt.year
    File['Week_day'] =File.Date.dt.weekday
    File['Day_name'] = File.Date.dt.weekday_name
    File['Week'] = File.Date.dt.weekofyear
    File['Returns'] = np.log(File.Close/File.Close.shift(1))
    return File
```

```python
cols = []
def display_all(df):
    with pd.option_context("display.max_rows", 14000, "display.max_columns", 1000): 
        display(df)

def feat(File):
    
    for i in range(0, len(File)):
        year = File.loc[i,'Year']
        week = File.loc[i,'Week']
        day = File.loc[i,'Week_day']
        lags = 5
        cols = []
        for lag in range(1, lags+1):
                   
            col = 'feat_%s' % (lag)
          
            try:
                File.loc[i, col] = File.loc[((File.Week_day == day)&(File.Week == week)&(File.Year== (year - lag)), 'Returns')].values[0]
            except:
                File.loc[i, col] = None
            
            
            cols.append(col)
            
    File.dropna(inplace=True)
    File[cols] = np.where(File[cols] > 0, 1, 0)
    File['Direction'] = np.where(File['Returns'] > 0, 1, -1)
    return display_all(File)
```

```python
data = clean(eu)
```

```python
feat(data)
```

```python
cols = ['feat_1', 'feat_2', 'feat_3', 'feat_4', 'feat_5']
data[cols + ['Direction']].head()
```

## *MACHINE LEARNING*

#### **Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
split = int(len(data) * 0.75)
train = data.iloc[:split].copy()
test = data.iloc[split:].copy()
```

```python
model = RandomForestClassifier(n_estimators=10,n_jobs=-1, min_samples_leaf=43)
#train[cols] = sc.fit_transform(train[cols])
#test[cols] = sc.transform(test[cols])
model.fit(train[cols], train['Direction'])
```

```python
accuracy_score(train['Direction'], model.predict(train[cols]))
```

```python
test['Position'] = model.predict(test[cols])
accuracy_score(test['Direction'], test['Position'])
```
